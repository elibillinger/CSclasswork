```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import sklearn.datasets
import datetime
import time
from pandas import Series, DataFrame
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

earthquakes = pd.read_table('database.csv', sep = ",")
pd.set_option('display.max_columns', 21)
```

```python


timestamp = []
for d, t in zip(earthquakes['Date'], earthquakes['Time']):
    try:
        ts = datetime.datetime.strptime(d+' '+t, '%m/%d/%Y %H:%M:%S')
        timestamp.append(time.mktime(ts.timetuple()))
    except ValueError:
        # print('ValueError')
        timestamp.append('ValueError')
timeStamp = pd.Series(timestamp)
earthquakes['Timestamp'] = timeStamp.values
final_data = earthquakes.drop(['Date', 'Time','ID','Type','Source','Root Mean Square','Depth Error','Depth Seismic Stations','Location Source','Magnitude Seismic Stations','Magnitude Source','Status','Horizontal Error','Magnitude Error','Horizontal Distance'], axis=1)
final_data = final_data[final_data.Timestamp != 'ValueError']
final_data
```

```python
features = ['Latitude','Longitude','Timestamp']
target = ['Magnitude','Depth']
X_train,X_test,y_train,y_test = train_test_split(final_data[features],final_data[target],test_size=0.2)
```

```python
model = RandomForestRegressor(n_estimators = 1000,max_depth=10)
#model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
